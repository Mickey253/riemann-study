```python
import json 
with open("src/application/data/sierpinski.txt", "r") as fdata:
    gtxt = fdata.read()
gtxt = gtxt.replace("\'", "\"")
gtxt[:100]
```

```python
gdata = json.loads(gtxt)
gdata
```

```python
for i,n in enumerate(gdata['nodes']):
    n["spherical"] = {
        "x": n["pos"][0],
        "y": n["pos"][1]
    }
    del n["pos"]
gdata['nodes'][0]
```

```python
gdata['links'] = gdata['edges']
del gdata['edges']
```

```python
gnum = 0
gtype = "s"
```

```python
for gnum in range(3):
    for gtype in ("s", "h", "e"):
        with open(f"src/application/data/{gtype}_group_{gnum}.json", "r") as fdata:
            gdata = json.load(fdata)

        for node in gdata['nodes']:
            tmp = node["spherical"]["y"]
            node["spherical"]["y"] = node["spherical"]["x"]
            node["spherical"]["x"] = tmp

        with open(f"src/application/data/{gtype}_group_{gnum}.json", "w") as fdata:
            json.dump(gdata,fdata,indent=4)
```
