```python
import json 
with open("src/application/data/sierpinski.txt", "r") as fdata:
    gtxt = fdata.read()
gtxt = gtxt.replace("\'", "\"")
gtxt[:100]
```

```python
gdata = json.loads(gtxt)
gdata
```

```python
for i,n in enumerate(gdata['nodes']):
    n["spherical"] = {
        "x": n["pos"][0],
        "y": n["pos"][1]
    }
    del n["pos"]
gdata['nodes'][0]
```

```python
gdata['links'] = gdata['edges']
del gdata['edges']
```

```python
with open("src/application/data/s_group_2.json", "w") as fdata:
    json.dump(gdata,fdata,indent=4)
```
